```python
from bs4 import BeautifulSoup
import importlib
import csv
import random
import itertools
from mqdq import line_analyzer as la
from mqdq import utils
import pandas as pd
import numpy as np
import scipy as sp
```

```python
ovam_clean = []
ovam_poems = []
for x in ['1','2','3']:
    with open("../OV-amo%s.xml" % x) as fh:
        s = BeautifulSoup(fh,"xml")
    ovam.append(utils.clean(s('line')))
    ovam_poems.append( [d('line') for d in s('division')] )
```

```python
def analyze_3fc(books):
    l_count=0
    no3c_count=0
    for bix, book in enumerate(books):
        for pix, poem in enumerate(book):
            for l in poem:
                if l['pattern']=='not scanned' or l['pattern']=='corrupt': continue
                if l['metre']!='H': continue
                l_count+=1
                if la.has_3rd_foot_caes(l, strict=True): continue
                no3c_count+=1
                if len(books) > 1:
                    print("%d:%d.%s" % (bix+1, pix+1, l['name']))
                else:
                    print("%d.%s" % (pix+1, l['name']))
                print(utils.txt(l, True))
                print()
    return (l_count, no3c_count)
            
```

```python
l_count, no3c_count = analyze_3fc(ovam_poems)
print("Amores: %d from %d (%.2f)%%" % (no3c_count, l_count, no3c_count/l_count*100))
```
